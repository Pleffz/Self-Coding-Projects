```python
# Libraries that may be used
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the file using pandas (source from kaggle)
data = pd.read_csv(r"C:\Users\jeffr\Downloads\EV_Population_Size_History_By_County.csv")
```

```python
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Calculate mean value for each year
summary_stats = df.groupby(df['Date'].dt.year)['Percent Electric Vehicles'].mean()
summary_stats = summary_stats.reset_index()  # Reset index to make year a column

# Plotting the bar plot
plt.figure(figsize=(10, 6))
plt.bar(summary_stats['Date'], summary_stats['Percent Electric Vehicles'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Mean Percentage')
plt.title('Mean Percentage by Year')
plt.xticks(summary_stats['Date'])
plt.grid(axis='y')
plt.show()
```

**Summary**
From the data gathered, we can conclude that there is significant growth in EV car usage.  From 2017-2020, less than 4% of cars being driven from the given data are EV cars. From 2021-2024, the amount of EV cars being driven are consistently above 5%.  There is growth in EV vehicles but there are too many cars that have been produced that are gas-powered and already owned to make a more significant change.  We can make more assumptions on the percentages based on the prices of the car, maintenance, and gasoline.  As gasoline prices continue to rise, more people will be geared to shift towards EV.

```python
plt.figure(figsize=(12, 8))

# Use seaborn to create an area plot
sns.lineplot(x='State', y='Percent Electric Vehicles', data=df, sort=False, linewidth=2)

# Customize labels and title
plt.xlabel('State', fontsize=12)
plt.ylabel('Percent Electric Vehicles', fontsize=16)  
plt.title('Percentage of Electric Vehicles by State', fontsize=16) 

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show plot
plt.tight_layout()
plt.show()
```

**Summary**
This plot shows the percentages of electric vehicles within a state (with the given dataset)
