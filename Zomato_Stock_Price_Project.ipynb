```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as pyplot
from matplotlib import style
from sklearn import linear_model
from sklearn.utils import shuffle

# Read the dataset using pandas
data = pd.read_csv(r'C:\Users\jeffr\Downloads\Zomato Dataset.csv')

# Choose the data you want to use
data = data[["Open", "High", "Low", "Close", "Adj Close", "Volume"]]

# Print the first 5 rows of the dataset
data.head()
```

```python
# Set up data for linear regression
predict = "High"
X = np.array(data.drop([predict], axis = 1))
y = np.array(data[predict])

# Print the accuracy of the linear regression model
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
linear = linear_model.LinearRegression()
linear.fit(X_train, y_train)
acc = linear.score(X_test, y_test)
print(acc)
```

```python
# Predict the 'High' Column
predictions = linear.predict(X_test).astype(int)
for i in range(len(predictions)):
    print(predictions[i], X_test[i], y_test[i])
```
